```python
import tensorflow as tf
import cv2
import numpy as np

# Create a function that takes a frame as input and outputs the aligned frame
def align_frame(frame1, frame2, sift):
    # Detect SIFT features in both frames
    kp1, des1 = sift.detectAndCompute(frame1, None)
    kp2, des2 = sift.detectAndCompute(frame2, None)

    # Match features between the frames
    bf = cv2.BFMatcher()
    matches = bf.match(des1, des2)

    # Get corresponding points in each frame
    points1 = np.float32([kp1[m.queryIdx].pt for m in matches]).reshape(-1,1,2)
    points2 = np.float32([kp2[m.trainIdx].pt for m in matches]).reshape(-1,1,2)

    # Find transformation matrix to align frames
    M, mask = cv2.findHomography(points1, points2, cv2.RANSAC, 5.0)

    # Apply transformation to frame1
    aligned_frame1 = cv2.warpPerspective(frame1, M, frame2.shape[1::-1])

    return aligned_frame1

# Create a SIFT object
sift = cv2.SIFT_create()

# Define the input signature for the function
input_signature = [
    tf.TensorSpec(shape=(None, None, 3), dtype=tf.uint8, name='frame1'),
    tf.TensorSpec(shape=(None, None, 3), dtype=tf.uint8, name='frame2')
]

# Wrap the function inside a tf.function to make it a TensorFlow graph
@tf.function(input_signature=input_signature)
def aligned_frame_fn(frame1, frame2):
    return tf.numpy_function(align_frame, [frame1, frame2, sift], tf.uint8)

# Save the model to disk as a tf.saved_model
#tf.saved_model.save(aligned_frame_fn, 'C:/Users/Bhupendra Singh/Documents/')

#tf.saved_model.save(aligned_frame_fn, 'C:/Users/Bhupendra Singh/Documents/Project_May_01/saved_model')
save_dir = 'C:/Users/Bhupendra Singh/Documents/Project_May_01/saved_model'
tf.saved_model.save(aligned_frame_fn, save_dir)
```
